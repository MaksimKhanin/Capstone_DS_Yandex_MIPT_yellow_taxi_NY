```python
%matplotlib inline

import pandas as pd
import numpy as np

from tqdm import tqdm_notebook as tqdm

from sklearn.linear_model import LinearRegression

from sklearn.cluster import KMeans

from sklearn.decomposition import PCA

from tqdm import tqdm_notebook as tqdm

#import plotly.graph_objects as go

import matplotlib.pyplot as plt
import matplotlib.cm as mplcm


import statsmodels.api as sm
from scipy import stats

import warnings

import pickle

from glob import glob
```

```python
df = pd.read_csv('taxi_agg.csv', index_col=0)
```

```python
regions = pd.read_csv('data_week2.csv', index_col=0).index # регионы отобрытнные на второй неделе (среднее к-во поездок в мае > 5)
df = df[df['bin_num'].isin(list(regions))]
```

```python
df.index = pd.to_datetime(df.index)
```

```python
df = df.set_index('Pickup_Hour')
df = df.pivot_table(values='trip_cnt', index=df.index, columns='bin_num', aggfunc='first')
df = df.fillna(0)
```

```python
df = df.loc[:'2016-05-01']
```

Поскольку для нас важна только форма ряда, нормализовывать мы будем каждый регион отедально (рассчитывая среднюю и дисперсию только этого ряда). Таким образом мы сохраняем форму ряда, но нивилируем его сдвиг по абсолютным значениям

```python
df_original = df.copy() # сохраняем для востановления ненормализованных значений
```

```python
for each_bin in df.columns:
    mean = df[each_bin].mean()
    std = df[each_bin].std()  
    df[each_bin] = (df[each_bin] - mean) / std
```

```python
df = df.T
```

Для оценки качества кластеризации будем использовать визуальный анализ объектов получившихся после PCA

```python
#%%capture
#l_score = []
# Для построения временных рядов по наиболее типичным значениям кластеров используем ненормализованные значения
Clustered_data = df_original.T.copy() 
n_num = range(3, 10)
cmap = mplcm.get_cmap('Spectral')
for each_n in tqdm(n_num):
    
    model = KMeans(n_clusters = each_n).fit(df.values)
    #l_score.append(davies_bouldin_score(df.values, model.labels_))
    
    pca = PCA(n_components = 2)
    pca.fit(df.values)
    xy = pca.transform(df.values)
    
    plt.figure(figsize=(15,7))
    scat = plt.scatter(x=xy[:,0], y=xy[:,1], marker='o', c = cmap(model.labels_/len(n_num)))
    plt.title('Number of Clusters is {}'.format(str(each_n)))
    #plt.legend(loc="upper left")
    plt.show()

    
    Clustered_data['cluster'] = model.labels_
    group_cluster = Clustered_data.groupby('cluster').mean()
    plt.figure(figsize=(15,7))
    for each_index in group_cluster.index:
        
        #plt.plot(group_cluster.columns.values, group_cluster.loc[each_index, :].values,  label='Cluster '+str(each_index), alpha = 0.8)
        group_cluster.loc[each_index, :].plot(label='Cluster '+str(each_index), alpha = 0.8, color = cmap(float(each_index)/len(n_num)))
     
    #plt.xlabel()
    plt.legend(loc="upper left")
    plt.gcf().autofmt_xdate()
    plt.xticks(rotation = 90)
    #fig.update_traces(marker_color = model.labels_, text = model.labels_)
    plt.show()
```

```python
ps = range(0, 3)
d = 1
qs = range(0, 3)
Ps = range(0, 2) 
D = 1
Qs = range(0, 2)
```

Создадим функцию обучения. Обучать будем на остатках (вместо передачи регрессионных признаков в exog)

```python
df_done = pd.read_csv('model_result.txt', sep=';', names=['Cluster', 'params', 'aic'])
df_done.head()
```

```python
from itertools import product
d_results = {}
for each_cluster in range(0, 4):
    parameters = product(ps, qs, Ps, Qs)
    parameters_list = list(parameters)
    
    ts = df_clustered['resid_'+str(each_cluster)].copy()
    
    #exog = reg_features
    
    results = []
    best_aic = float("inf")
    warnings.filterwarnings('ignore')

    for param in parameters_list:
        #try except нужен, потому что на некоторых наборах параметров модель не обучается
    #     if param in known_params:
    #         continue
    
        if df_done[(df_done['Cluster'] == each_cluster) & (df_done['params'] == str(param))].empty == False:
             continue
    
        try:
            print('edu with ', param)
            model=sm.tsa.statespace.SARIMAX(ts, 
                                            order=(param[0], d, param[1]), 
                                            seasonal_order=(param[2], D, param[3], 24)).fit(disp=-1)
        #выводим параметры, на которых модель не обучается и переходим к следующему набору
        except:
            print('wrong parameters:', param)
            
            with open('model_result.txt', 'a') as f:
                f.write(str(each_cluster)+';'+str(param)+';'+'NULL'+'\n')
            
            continue
        aic = model.aic
        #сохраняем лучшую модель, aic, параметры 
        if aic < best_aic:
            #best_model = model
            best_aic = aic
            best_param = param
        results.append([param, model.aic])
        print('done edu with params =', param, model.aic)
        
        with open('model_result.txt', 'a') as f:
            f.write(str(each_cluster)+';'+str(param)+';'+str(model.aic)+'\n')
        
    warnings.filterwarnings('default')
    
    d_results[each_cluster] = results
```
